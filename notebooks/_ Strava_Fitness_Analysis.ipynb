```python
hourly_steps["ActivityHour"] = pd.to_datetime(hourly_steps["ActivityHour"])

hourly_steps["Hour"] = hourly_steps["ActivityHour"].dt.hour
```

```python
hourly_avg = (
    hourly_steps.groupby("Hour")["StepTotal"]
    .mean()
)

plt.figure(figsize=(12,5))

plt.plot(hourly_avg.index, hourly_avg.values, marker="o")

plt.title("Average Steps by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Steps")

plt.grid(True)

plt.show()
```

```python
hourly_calories["ActivityHour"] = pd.to_datetime(hourly_calories["ActivityHour"])

hourly_calories["Hour"] = hourly_calories["ActivityHour"].dt.hour

hourly_cal_avg = (
    hourly_calories.groupby("Hour")["Calories"]
    .mean()
)

plt.figure(figsize=(12,5))

plt.plot(hourly_cal_avg.index, hourly_cal_avg.values, marker="o")

plt.title("Average Calories Burned by Hour")
plt.xlabel("Hour")
plt.ylabel("Calories")

plt.grid(True)

plt.show()
```

```python
def activity_level(steps):
    if steps < 5000:
        return "Sedentary"
    elif steps < 7500:
        return "Low Active"
    elif steps < 10000:
        return "Moderately Active"
    elif steps < 12500:
        return "Active"
    else:
        return "Highly Active"

daily_activity["ActivityLevel"] = daily_activity["TotalSteps"].apply(activity_level)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=daily_activity,
    x="ActivityLevel",
    order=[
        "Sedentary",
        "Low Active",
        "Moderately Active",
        "Active",
        "Highly Active"
    ]
)

plt.title("User Activity Level Distribution")
plt.xlabel("Activity Level")
plt.ylabel("Number of Records")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=daily_activity,
    x="ActivityLevel",
    y="Calories",
    estimator=np.mean,
    order=[
        "Sedentary",
        "Low Active",
        "Moderately Active",
        "Active",
        "Highly Active"
    ]
)

plt.title("Average Calories Burned by Activity Level")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=daily_activity,
    x="ActivityLevel",
    y="TotalDistance",
    estimator=np.mean,
    order=[
        "Sedentary",
        "Low Active",
        "Moderately Active",
        "Active",
        "Highly Active"
    ]
)

plt.title("Average Distance by Activity Level")

plt.show()
```
